## 1. What is the difference between pivot and pivot_tables?

pivot: This is used to reshape data based on column values. It takes three arguments: index, columns, and values.


pivot_table: This is a generalization of pivot that allows for aggregation of numeric data when there are duplicate entries.
    



## 2. What is the difference between wide & long tables?
Wide Table: Data is represented with multiple variables (columns) as different columns.

Long Table (or Tidy Data): Data is represented with multiple variables (columns) as different rows.

## 3. What pandas functions can you use to change from wide to long and vice versa?

For changing from wide to long format: pd.melt()

For changing from long to wide format: pd.pivot() or pd.pivot_table()

## 4. You are a manager of a store outlet in your city, with 5 sales representatives, 
create a sale inventory DataFrame for the 1st quarter of 2024.

Columns: Name, sex, product, total_sales_number, total_sales_cost (add other columns as you wish)


```python
np.random.seed(100)
data = {
    'Name': ['John', 'Alice', 'Bob', 'Charlie', 'Emma'] * 15,
    'Sex': ['M', 'F', 'M', 'M', 'F'] * 15,
    'Product': np.random.choice(['A', 'B', 'C', 'D', 'E'], size=75),
    'Month': np.random.randint(1, 4, size=75),
    'Total_Sales_Number': np.random.randint(10, 100, size=75),
    'Total_Sales_Cost': np.random.randint(50000, 200000, size=75)
}

# Create the sale inventory DataFrame
sale_inventory_df = pd.DataFrame(data)
sale_inventory_df
```

```python
# Convert 'Month' column to datetime format
sale_inventory_df['date'] = pd.to_datetime(sale_inventory_df['Month'].astype(str) + '/1/2024')
sale_inventory_df=sale_inventory_df.sort_values(by='date')
sale_inventory_df
```

```python
sale_inventory_df['Month']=sale_inventory_df['Month'].replace({1:'January', 2:'Feburary', 3:'March'})
sale_inventory_df
```

### 4a Get the sum of the total sales


```python
sale_inventory_df['Total_Sales_Cost'].sum()
```

### 4b Calculate the cost of each product and assign to a new column


```python
sale_inventory_df['Cost_Per_Product'] = sale_inventory_df['Total_Sales_Cost'] / sale_inventory_df['Total_Sales_Number']
sale_inventory_df['Cost_Per_Product']=sale_inventory_df['Cost_Per_Product'].round(2)
sale_inventory_df
```

### 4c Get the average sale for each sale rep


```python
sale_inventory_df.groupby('Name')['Total_Sales_Cost'].mean()
```

```python
sale_inventory_df.groupby('Name')['Total_Sales_Cost'].mean().plot(kind='bar')
```

```python
# Average sale of each product
sale_inventory_df.groupby('Product')['Total_Sales_Cost'].mean().plot(kind='bar')
```

## 1. What is the difference between Series & DataFrame


**Series:** A Series is a one-dimensional array-like object that can hold any data type (integers, strings, floating-point numbers, etc.). It is similar to a column in a table or a single column in a DataFrame. Each value in a Series is associated with an index.

**DataFrame:** A DataFrame is a two-dimensional, size-mutable, and heterogeneous tabular data structure with labeled axes (rows and columns). It is similar to a table in a database or a spreadsheet. Each column in a DataFrame is a Series.

## 1. What's the Difference Between merge, join, and concatenate in Pandas?


In Pandas, merge, join, and concatenate are functions used to combine data from multiple DataFrames. While they serve similar purposes, they have different use cases and methods of operation.

**merge**

The merge function combines DataFrames based on a key column or multiple key columns. It is similar to SQL joins (inner, outer, left, and right joins).
- Combines DataFrames using common columns.
- Supports different types of joins (inner, outer, left, right).
- Syntax: pd.merge(left, right, how='inner', on='key')

**join**

The join function is used to join two DataFrames on their index or a key column. It is particularly useful for combining DataFrames with a similar index.

- Joins DataFrames using their index or a specified key column.
- Supports different types of joins (inner, outer, left, right).
- Syntax: df1.join(df2, how='left', on='key')

**concatenate**

The concatenate (or concat) function concatenates DataFrames along a particular axis (rows or columns). It stacks DataFrames either vertically (along rows) or horizontally (along columns).

- Concatenates DataFrames along a specified axis (0 for rows, 1 for columns).
- Does not require common columns or indexes.
- Syntax: pd.concat([df1, df2], axis=0)

## 2. Explain the Concept of Inner and Outer Join in DataFrame Merging

Inner Join: Returns only the rows where the merge/intersection of the keys (columns) exist in both DataFrames.


Outer Join: Returns all rows from both DataFrames, filling in missing values with NaN where keys do not match.

Grouping, aggregating, reshaping, pivot table, melt dataframs with Pandas


## 1. What is the difference between pivot and pivot_tables?

`pivot`

This is used to reshape data based on column values. Used to transform or reorganize data to give a new presentation of the data. Cannot handle duplicate entries for the same index/column pair. It takes three arguments: index, columns, and values.

Syntax: DataFrame.pivot(index=None, columns=None, values=None)



`pivot_table`

This is a generalization of pivot that allows for aggregation (mean, sum etc) of numeric data when there are duplicate entries. Creates a spreadsheet-style pivot table as a DataFrame. Allows for more complex aggregation and handling of duplicate values. Provides an aggregation function to combine duplicate entries, which makes it more versatile.
    
Syntax: DataFrame.pivot_table(index=None, columns=None, values=None, aggfunc='mean')


## 2. What is the difference between wide & long tables?

Wide Table: Data is represented with multiple variables (columns) as different columns. Each variable is a separate column.


| date         | New York   | Los Angeles
| ---           | ---       | --- |
| 2021-01-01   | 30         |40   |
| 2021-01-02   |35          |45   |


Long Table (or Tidy Data): Data is represented with multiple variables (columns) as different rows. Each observation is a row; the values of different variables are stored in a single column.


|date         |city          |temperature
| ---           | ---       | --- |
|2021-01-01   |New York      |30
|  2021-01-01   |Los Angeles   |40
|  2021-01-02   |New York      |35
|  2021-01-02   |Los Angeles   |45


## 3. What pandas functions can you use to change from wide to long and vice versa?

For changing from wide to long format: pd.melt()


For changing from long to wide format: pd.pivot() or pd.pivot_table()

## 3. Calculate rolling mean, sum, or other statistics over a specified window of time.

what is rolling mean?

In time series analysis, "rolling" (or moving) calculations involve creating a series of statistics over a specified window of time, which moves step-by-step along the data series. This can be particularly useful for smoothing out short-term fluctuations and highlighting longer-term trends or cycles



A rolling mean (also known as a moving average) calculates the average of the last 
𝑛
n data points for each point in the series. As the window moves forward, it includes the next data point and excludes the oldest data point in the previous window.

```python
rolling_mean = df['meantemp'].rolling(window=7).mean()
rolling_mean
```

```python
rolling_sum = df['meantemp'].rolling(window=7).sum()
rolling_sum
```
